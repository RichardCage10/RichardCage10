Muestra un DataFrame con los datos de la siguiente tabla:

```
  Mes     Ingresos     Gastos

  Febrero 17600        15400

```



````python
Datos = {' Mes ': ['Enero','Febrero','Marzo','Abril'],'Ingresos': [21500,17600,25000,34800],'Gastos':[17000,15400,19200,22500]}
df = pd.DataFrame(Datos)
df
````

Escribe una funcion que reciba el DataFrame del ejercicio anterior y devuelva el beneficio total de cada mes (Ingresos - Gastos).


````python
def Beneficio(Datos):
  Datos['Beneficio'] = Datos['Ingresos'] - Datos['Gastos']
  return Datos
````

````python
#1. Cargar DataSet
titanic = pd.read_csv('/content/drive/MyDrive/datasets/titanic.csv', index_col=0)
titanic.head()
````

````python
#2. Mostramos dimensiones
print("Dimensiones : ", titanic.shape)
print("Numero de datos : ", titanic.size)
print("Nombres de columnas : ", titanic.columns)
print("Nombres de filas : ", titanic.index)
print("Tipos de datos : \n", titanic.dtypes)
````

````python
#3. Pasajero 121
titanic.loc[121]
````

````python
#4. Filas impares del df
titanic_impares = titanic.iloc[range(0,titanic.shape[0],2)]
titanic_impares.head(15)
````

````python
#5. Porcentaje de supervivientes y de muertes
porcentaje_titanic=titanic['Survived'].value_counts()/titanic['Survived'].count()*100
porcentaje_titanic
````

````python
#6. Segunda clase
pasajeros_segunda=titanic[titanic['Pclass']==2]['Name'].sort_values()
pasajeros_segunda.head(5)
````

````python
#7. Eliminar nulos de edad
titanic_edadesnonulas=titanic.dropna(subset=['Age'])
titanic_edadesnonulas
````

````python
#8. Porcentaje de muertes por cada clase
titanic_por_clase=titanic.groupby('Pclass')['Survived'].value_counts(normalize=True)*100
titanic_por_clase
````

````python
#9. Edad media por clase
titanic.groupby(['Pclass','Sex'])['Age'].mean()
````

````python
#10. Menor de edad
titanic['Menor']=titanic['Age']<18
titanic.head(10)
````

````python
#11. Menores y Mayores
titanic.groupby(['Pclass','Menor'])['Survived'].value_counts(normalize=True)*100
````
